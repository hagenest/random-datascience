```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('data.xlsx', sheet_name='Sheet1')
```

```python
plt.figure(figsize=(25.6, 19.2))
plt.scatter(df['YouTube-Views'], df['Public Vote'])
plt.xlabel('YouTube-Views')
plt.ylabel('Public Vote')
plt.title('Scatter Plot of YouTube-Views vs Public Votes')
for index, row in df.iterrows():
    plt.text(row['YouTube-Views'], row['Public Vote'], row['Country'])
z = np.polyfit(df['YouTube-Views'], df['Public Vote'], 1)
p = np.poly1d(z)
plt.plot(df['YouTube-Views'], p(df['YouTube-Views']))
plt.show()
```
